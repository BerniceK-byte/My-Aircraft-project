### Handling missing data


For ccurate analysis of data ,It is important to work with the correct data.Missing values and null values can bring a lot of inaccuracy when doing analysis and hinder the making of informed decisions.It is therefore necessary to find columns with missing data,duplicated data as well as deleting columns that may not be helpful in analysis

```python
missing_percentage = df2.isna().sum()/df2.shape[0]*100
```

```python

def drop_columns_with_high_missing_values(df2, threshold=0.5):
    # Calculate the percentage of missing values for each column
    missing_percentage = df2.isnull().mean()

    # Identify columns that exceed the threshold of missing values
    columns_to_drop = missing_percentage[missing_percentage > threshold].index  

    # Check if any columns need to be dropped
    if len(columns_to_drop) > 0:
        df2.drop(columns=columns_to_drop, inplace=True)
        print(f"Dropped columns with > {threshold * 100}% missing values: {columns_to_drop.tolist()}")
    else:
        print("No columns exceeded the missing value threshold.")

    return df

# Call the function to drop columns with more than 50% missing values
df2_cleaned = drop_columns_with_high_missing_values(df2, threshold=0.5)

# Print cleaned DataFrame
print("\nCleaned DataFrame:")
print(df2_cleaned)
```

```python
# Check the first few rows to understand the data
print(df2.head())

# Check for missing values
missing_values = df2.isnull().sum()
print(f"Missing values per column:\n{missing_values}")

# Fill missing values for numeric columns with the median (use mean if preferred)
for column in df2.select_dtypes(include=['float64', 'int64']).columns:
    median_value = df2[column].median()
    df2[column].fillna(median_value, inplace=True)
    print(f"Filled missing values in column '{column}' with median: {median_value}")

# Fill missing values for categorical columns with the mode
for column in df2.select_dtypes(include=['object']).columns:
    mode_value = df2[column].mode()
    df2[column].fillna(mode_value, inplace=True)
    print(f"Filled missing values in column '{column}' with mode: {mode_value}")

# Check again for missing values
print("\nMissing values after handling:")
print(df2.isnull().sum())
```
